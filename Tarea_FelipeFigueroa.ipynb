# Question 1: Use yfinance to Extract Stock Data

Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function. Upload a screenshot of the results and code from the beginning of Question 1 to the results below.

```python
import yfinance as yf

# Definir el ticker de Tesla
ticker_tesla = 'TSLA'

# Obtener el objeto Ticker de Tesla
tesla = yf.Ticker(ticker_tesla)

# Obtener los datos históricos de las acciones
tesla_data = tesla.history(period="1y")  # Datos de los últimos 12 meses

# Restablecer el índice
tesla_data = tesla_data.reset_index()

# Guardar los datos en un archivo CSV
tesla_data.to_csv("tesla_stock_data.csv", index=False)

# Mostrar los primeros cinco datos históricos
print(tesla_data.head())
```

# Question 2: Use Webscraping to Extract Tesla Revenue Data

Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de MacroTrends para los ingresos de Tesla
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Agregar cabeceras para simular un navegador
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

# Realizar la solicitud HTTP
response = requests.get(url, headers=headers)

if response.status_code == 200:
    print("Conexión exitosa")
    
    # Analizar el HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Buscar la tabla que contiene los ingresos de Tesla
    table = soup.find("table", {"class": "historical_data_table table"})
    
    # Obtener las filas de la tabla
    rows = table.find_all("tr")
    
    # Crear listas para las fechas y los ingresos
    dates = []
    revenues = []
    
    # Iterar sobre las filas y extraer los datos
    for row in rows[1:]:  # Saltar la primera fila (encabezados)
        cols = row.find_all("td")
        if len(cols) > 1:
            dates.append(cols[0].text.strip())  # Año
            revenues.append(cols[1].text.strip())  # Ingreso
    
    # Crear un DataFrame con los datos extraídos
    data = {
        "Fecha": dates,
        "Ingresos (Millones de USD)": revenues
    }
    tesla_revenue = pd.DataFrame(data)
    
    # Mostrar las últimas 5 filas utilizando tail
    print("\nÚltimas 5 filas de los ingresos de Tesla:")
    print(tesla_revenue.tail())

else:
    print("Error al conectarse:", response.status_code)
```

# Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Definir el ticker de Tesla
ticker_tesla = 'TSLA'

# Obtener el objeto Ticker de Tesla
tesla = yf.Ticker(ticker_tesla)

# Obtener los datos históricos de las acciones
tesla_data = tesla.history(period="1y")  # Solicitar datos de los últimos 12 meses

# Función para graficar los datos de las acciones
def make_graph(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Precio de Cierre', color='blue')  # Graficar el precio de cierre
    plt.title('Acciones de Tesla (TSLA) - Precio de Cierre')  # Título del gráfico
    plt.xlabel('Fecha')  # Etiqueta en el eje X
    plt.ylabel('Precio de Cierre (USD)')  # Etiqueta en el eje Y
    plt.grid(True)  # Mostrar la cuadrícula
    plt.legend()  # Mostrar la leyenda
    plt.xticks(rotation=45)  # Rotar las etiquetas de las fechas
    plt.tight_layout()  # Ajustar el espacio para evitar el solapamiento
    plt.show()

# Llamar a la función make_graph para graficar los datos de Tesla
make_graph(tesla_data)
```
